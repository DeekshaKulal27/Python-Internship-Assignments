```python
messages = ["Hi", "Welcome to the platform", "OK"]
```

```python
print("message length analysis")
for message in messages:
    length = len(message)
    print(f"Message: '{message}'")
    print("Length:", length)
    
    if length > 10:
        print("Status: flagged\n")
    else:
        print("Status: valid message\n")
```
