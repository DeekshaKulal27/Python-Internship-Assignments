```python
marks = [45, 78, 90, 33, 60]
```

```python
pass_count = 0
fail_count = 0
```

```python
print(" Online Examination Result ")
for i in range(len(marks)):
    print(f"Student {i+1} Marks: {marks[i]}")
    
    if marks[i] >= 50:
        print("Result: Pass\n")
        pass_count += 1
    else:
        print("Result: Fail\n")
        fail_count += 1
print("Final Summary ")
print("Total Students:", len(marks))
print("Total Passed:", pass_count)
print("Total Failed:", fail_count)
```
